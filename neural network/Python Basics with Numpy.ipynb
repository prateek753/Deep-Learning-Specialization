```python
#sigmoid function or logistic function
#In deep learning we mostly use matrices and vectors, thus, we'll use numpy
def sigmoid(x) :
    return 1/(1+np.exp(-x))

# np.exp(x) works for any np.array(x) and applies the exponential function to every coordinate
# Note that derivative of sigmoid is s*(1-s) , if s = sigmoid(x)
```

```python
# Reshaping Arrays
# X.shape is used to get the shape/dimension of matrix/vector X.
# X.reshape(...) is used to reshape dimension of X into some other dimension
#In computers images are represented as 3D arrays of shape (length,width,height). But, in DL when you 
#read images as input you need to convert the images into a vector of shape(length*width*height , 1).
#Thus, in other words we need to reshape 3D array into 1D vector
def image2vector(image):
    v = image.reshape(image.shape[0]*image.shape[1]*image.shape[2] , 1)
    return v

# image = a numpy array of shape (length, height, depth)
# v = a vector of shape (length*height*depth, 1)
```

## Normalizing Rows


```python
def normalize_rows(x):
    x_norm = np.linalg.norm(x,axis=1,keepdims=True)
    x = x/x_norm
    return x
    
# x is a numpy matrix of shape (n, m)
```

### Softmax Function


```python
def softmax(x):
    x_exp = np.exp(x)
    x_sum = np.sum(x_exp , axis = 1 , keepdims = True)
    s = x_exp/x_sum    
    return s
```
